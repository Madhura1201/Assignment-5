```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import (classification_report, confusion_matrix,
                             accuracy_score, ConfusionMatrixDisplay)
from sklearn.preprocessing import StandardScaler
```

```python
# ─── Cell 2: Load Dataset ─────────────────────────────────────────────────────
# Fix: Use raw string (r prefix) or double backslashes or forward slashes to avoid escape sequence issues
df = pd.read_csv(r"C:\Users\HP\Downloads\Mktg_Campaign.csv")
# Alternative solutions:
# df = pd.read_csv("C:\\Users\\HP\\Downloads\\Mktg_Campaign.csv")  # Double backslashes
# df = pd.read_csv("C:/Users/HP/Downloads/Mktg_Campaign.csv")      # Forward slashes
print("Shape:", df.shape)
df.head()
```

```python
# ─── Cell 3: Explore Data ─────────────────────────────────────────────────────
print("Column Names:\n", df.columns.tolist())
print("\nData Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())
print("\nTarget (Response) distribution:\n", df['Response'].value_counts())
df.describe()
```

```python
# ─── Cell 4: Handle Missing Values ───────────────────────────────────────────
df['Income'].fillna(df['Income'].median(), inplace=True)
print("Missing values after fix:\n", df.isnull().sum())
```

```python
# ─── Cell 5: One-Hot Encode Categorical Columns ───────────────────────────────
df_encoded = pd.get_dummies(df, columns=['Education', 'Marital_Status'])
print("Encoded columns:\n", df_encoded.columns.tolist())
df_encoded.head()
```

```python
# ─── Cell 6: Define Features (X) and Target (y) ──────────────────────────────
# Target column is 'Response'
X = df_encoded.drop('Response', axis=1)
y = df_encoded['Response']
print("X shape:", X.shape)
print("y shape:", y.shape)
print("Target column confirmed:", 'Response' in df_encoded.columns)
```

```python
# ─── Cell 7: Train/Test Split (70/30) ────────────────────────────────────────
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42)
print("Train size:", X_train.shape)
print("Test size:", X_test.shape)
```

```python
# ─── Cell 8: Scale Features (for Logistic Regression) ────────────────────────
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)
```

```python
# ─── Cell 9: Logistic Regression Model ───────────────────────────────────────
lr_model = LogisticRegression(max_iter=1000, random_state=42)
lr_model.fit(X_train_scaled, y_train)
lr_pred = lr_model.predict(X_test_scaled)

print("=== Logistic Regression Results ===")
print("Accuracy:", round(accuracy_score(y_test, lr_pred), 4))
print("\nClassification Report:")
print(classification_report(y_test, lr_pred))
```
